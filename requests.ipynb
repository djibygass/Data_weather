## Total Records for each city

```python
list(collection.aggregate(
    [{
        "$group" :
            {
                "_id" : "$location",
                "Total Subjects" : {"$sum" : 1}
            }
    }]
))
```

## Afficher les 10 plus récente donnée de miami

```python
list(collection.aggregate([
    {
        "$match" :{
            "location" : "Miami"
        }
    },{
        "$sort" :{
            "date_time" : pymongo.DESCENDING
        }
    },{
        "$limit" : 10
    }
]))
```

# transformer tempC (string) en (Double)

```python
collection.update_many({},[
    {
        "$set" : {
            "tempC" : {
                "$toDouble" : "$tempC"
            }
        }
    }
])
```

## Calculer la moyenne des température pour chaque ville et trier de la ville la plus froide à la plus chaude

```python
list(collection.aggregate(
    [{
        "$group" :
            {
                "_id" : "$location",
                "Temperature avg" : {"$avg" : "$tempC"}
            }
    },{
         '$sort' : {
             'Temperature avg' : 1
         }
    }
    ]
))
```

## Redefinition

## Les villes où le ressenti a atteint 0 le 01-01-2022

```python
list(collection.aggregate([
    {
        '$match' : {
            'FeelsLikeC' : '0',
            'date_time' : '2022-01-01'
        }
    },
    {
        '$group' : {
            '_id' : '$location'
        }
    }
]))
```
